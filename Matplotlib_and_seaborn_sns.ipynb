```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
y=2*x
y
```

```python
#Line plot
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y=2*x
y2=3*x
```

```python
plt.plot(x,y,color='g',linestyle=":",linewidth=2)
plt.plot(x,y2,color='r',linestyle="-",linewidth=3)
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='g',linestyle=":",linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle="-",linewidth=3)
plt.grid(True)
plt.show()
```

```python
#Bar plot
student={'Bob':87,'Matt':56,'Sam':27}
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
#Horizontal Bar plot.
plt.barh(names,values)
plt.show()
```

```python
#Scatter plt.
x=[10,20,30,40,50]
a=[6,7,8,9,10]
plt.scatter(x,a)
plt.show()
```

```python
plt.scatter(x,a,marker="*",c='g',s=100)
plt.show()
```

```python
x=[10,20,30,40,50]
a=[6,7,8,9,10]
b=[2,3,4,5,6]
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c='g',s=100)
plt.subplot(1,2,1)
plt.scatter(x,b,marker=".",c='r',s=200)
plt.show()
```

```python
#Histogram
data=[1,3,4,4,9,5,5,5,6,7]
plt.hist(data)
plt.show()
```

```python
plt.hist(data,color='g',bins=4)
plt.show()
```

```python
#Seaborn Line Plot.
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

```python
#Seaborn Bar plot.
sns.set(style='whitegrid')
pokemon=pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
sns.barplot(x='is_legendary',y='speed',hue='generation',data=pokemon,palette='Blues_d')#or rocket or vlag#,color='green' for fixed color)

plt.show()
```

```python
#Scatter plt.
iris=pd.read_csv('Iris.csv')
iris.head()
```

```python
sns.scatterplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,hue="Species",style="Species")
```

```python
#Histograms/Distplot
diamonds=pd.read_csv('diamonds.csv.zip')
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False,color='green')
```

```python
sns.distplot(diamonds['price'],bins=14,kde=False)
```

```python
sns.distplot(diamonds['price'],color='orange',vertical=True)
```

```python
#Joint Plot
iris=pd.read_csv('Iris.csv')
iris.head()
```

```python
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,color='olive')
```

```python
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,color='olive',kind='reg')
```

```python
#Box Plot.
churn=pd.read_csv('Churn_Modelling.csv')
churn.head()
```

```python
sns.boxplot(x='Exited',y='Tenure',data=churn,palette='Set1',linewidth=3)
#sns.boxplot(x='Credit',y='Tenure',data=churn,order=['Two year','Month-to-Month','One year'])
```

```python
#Seaborn Pair Plot
df=pd.read_csv('Iris.csv')
sns.pairplot(df, hue="Species")
plt.show()
```
